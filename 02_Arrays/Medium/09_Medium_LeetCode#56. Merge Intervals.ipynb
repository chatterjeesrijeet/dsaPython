```python
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        
        #---------------------update slots based on windows ------------
        
        slots = [-1]*10001
        solution = []
        
        for ele in intervals:
            i = ele[0]+1
            while i <= ele[1]:
                slots[i] = 1
                i += 1
        # print(slots)
        
        #----------------based on slots create the windows ----------------
        i = 0
        while i <= 10000:  
            
            if slots[i] == 1:
                j = i
                while slots[j] == 1:
                    j += 1
                solution.append([i-1,j-1])
                i = j
            else:
                i += 1
        
        #-------------------- inlusion of [i.i] intervals --------
        for ele in intervals:
            if ele[0]==ele[1]:
                if slots[ele[0]] == 1 or slots[ele[0]+1]==1 or (ele in solution):
                    pass
                else:
                    solution.append(ele)
                
        return solution
```
