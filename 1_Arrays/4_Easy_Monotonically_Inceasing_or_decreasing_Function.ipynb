```python
class Solution:
    def isMonotonic(self, nums: List[int]) -> bool:
        size = len(nums)
        i = 0
        if size <= 2 :
            return True
        else:
            # i = 0 
            while i < size-1 and nums[i] == nums[i+1]:
                i += 1
                if i == size-2:
                    return True
            if nums[i]- nums[i+1] > 0 :
                dec = True
            else:
                dec = False
        
        if dec == True :
            while i <= len(nums)-2:
                if nums[i]-nums[i+1] < 0 :
                    return False
                i += 1
        else:
            while i <= len(nums)-2:
                if nums[i]-nums[i+1] > 0 :
                    return False
                i += 1
        return True
```

```python
def isMonotonic(nums):
    size = len(nums)
    i = 0
    if size <= 2 :
        return True

    upward,downward = True, True
    for i in range(len(nums)-1):

        if nums[i] > nums[i+1]:
            upward = False
        elif nums[i] < nums[i+1]:
            downward = False

        if upward == False and downward == False:
            return False

    return True
```
