    our solution : 
    
              push : O(1)
              pop :  O(n)

```python
class MyQueue:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.s1 = []
        self.s2 = []
        

    def push(self, x: int) -> None:
        """
        Push element x to the back of queue.
        """
        self.s1.append(x)
        

    def pop(self) -> int:
        """
        Removes the element from in front of queue and returns that element.
        """
        while len(self.s1) > 1:
            temp = self.s1.pop()
            self.s2.append(temp)
            
        poppedele = self.s1.pop()
        
        while len(self.s2) > 0:
            temp = self.s2.pop()
            self.s1.append(temp)
            
        return poppedele

    def peek(self) -> int:
        """
        Get the front element.
        """
        while len(self.s1) > 1:
            temp = self.s1.pop()
            self.s2.append(temp)
            
        topele = self.s1[0]
        
        while len(self.s2) > 0:
            temp = self.s2.pop()
            self.s1.append(temp)
            
        return topele
    

    def empty(self) -> bool:
        """
        Returns whether the queue is empty.
        """
        return len(self.s1)==0
```

### A faster Solution : 

     The new faster approach : 
    
       >> Old Appraoch :  For pop() :
       
             1. pop from s1 and push to s2 
             2. then pop from s2 
             3. push elements back to s1 
            
       >> New Approach : No ned to push it back to s1, 
            
             1. we will keep popping from s2 only as long as its non empty 
        
             2. When its empty again we will pop from s1 and push to s2 and then start popping from s2 again.
        

     This way push and pop both takes O(1) , push = O(1) and pop() : amortized O(1)

```python
class MyQueue:

    def __init__(self):
        self.s1 = []
        self.s2 = []        

    def push(self, x: int) -> None:
        self.s1.append(x)
        

    def pop(self) -> int:
        
        if self.empty():
            return -1
        
        if len(self.s2)==0:
            while len(self.s1) > 0:
                temp = self.s1.pop()
                self.s2.append(temp)
        return self.s2.pop()

    
    '''Peek : The front element is kept in constant 
              memory and is modified when we push an element. 
              
              When s2 is not empty, front element is 
              positioned on the top of s2'''
    
    def peek(self) -> int:
        if self.empty():
            return -1
        
        if len(self.s2)!= 0:
            return self.s2[-1]
        return self.s1[0]
    

    def empty(self) -> bool:
        """
        Returns whether the queue is empty.
        """
        if len(self.s1) == 0 and  len(self.s2)==0:
            return True
        else:
            return False 
```
