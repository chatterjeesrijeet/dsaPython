```python
class Solution:
    def trap(self, height: List[int]) -> int:
        
        n = len(height)
        
        if n < 3:
            return 0
        
        runninglmax = [0 for i in range(n)]
        runningrmax = [0 for i in range(n)]
        
        for i in range(1,n):
            runninglmax[i] = max(height[i-1],runninglmax[i-1])
            # runningrmax[n-1-i] = max(height[n-i],runningrmax[n-i])
        
        print(f"runninglmax : {runninglmax}")
        
        for i in reversed(range(0,n-1)):
            runningrmax[i] = max(height[i+1],runningrmax[i+1])
        print(f"runningrmax : {runningrmax}")    
        
        tappedWater = 0 
        
        for i in range(n):
            diff = min(runninglmax[i],runningrmax[i]) - height[i]
            if diff > 0:
                tappedWater += diff
            
        return tappedWater
            
```

A little modification to make it faster :
    
    
    compute the leftrunningmax and tappedwater@i th index in the same loop as right running is already computed 

```python
class Solution:
    def trap(self, height: List[int]) -> int:
        
        n = len(height)
        
        if n < 3:
            return 0
        
        runninglmax = [0]
        runningrmax = [0 for i in range(n)]
                
        for i in reversed(range(0,n-1)):
            runningrmax[i] = max(height[i+1],runningrmax[i+1])
        
        tappedWater = 0 
        
        for i in range(1,n):
            runninglmax.append(max(height[i-1],runninglmax[i-1]))
            # runningrmax[n-1-i] = max(height[n-i],runningrmax[n-i])
            
            diff = min(runninglmax[i],runningrmax[i]) - height[i]
            if diff > 0:
                tappedWater += diff
        
        return tappedWater
            
```
