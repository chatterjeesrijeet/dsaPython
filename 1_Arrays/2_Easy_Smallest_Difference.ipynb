       Appraoch 1 : 
             >> Check every possible pairs using two arrays and keep the smallest difference one 
    
       Time Complexity = O(n*m)
       Space Complexity = O(1)

```python
import sys

def smallestDiff_V1(arr1,arr2):
    sd = sys.maxsize
    solutionPair = []
    for i in range(len(arr1)):
        for j in range(len(arr2)):
            if sd > abs(arr1[i]-arr2[j]):
                sd = abs(arr1[i]-arr2[j])
                solutionPair = [arr1[i],arr2[j]]
    return solutionPair
```

```python
arr1 = [-1,5,10,20,28,3]
arr2 = [26,134,135,15,17]
solutionPair = smallestDiff_V1(arr1,arr2)
print(solutionPair)
```

    Appraoch 2 : 
    
          >> sort the arrays 
          >> start compairing with two pointers to two arrays and try to come as close as possible 
    
    Time Complexity = O(nlogn + mlogm + m + n)
    Space Complexity = O(1)

```python
def smallestDiff_V2(arr1,arr2):
    
    arr1.sort()
    arr2.sort()
    
    i = 0
    j = 0
    
    minDiff = sys.maxsize
    
    while (i <= len(arr1)-1 and j <= len(arr2)-1 ) and minDiff != 0:
#         print(i,j)
        currDiff = abs(arr1[i] - arr2[j])
        
        if currDiff < minDiff:
            minDiff = currDiff
            solution = [arr1[i],arr2[j]]
        
        if arr1[i] < arr2[j]:
            i += 1
        
        elif arr2[j] < arr1[i]:
            j += 1
        else:
            return solution
    
    return solution    
```

```python
arr1 = [-1,5,10,20,28,3]
arr2 = [26,134,135,15,17,7]
solutionPair = smallestDiff_V2(arr1,arr2)
print(solutionPair)
```
