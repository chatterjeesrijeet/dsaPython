### Approach 1 

```python
class Solution:
    def kthGrammar(self, N: int, k: int) -> int:
        
        if N == 1:
            return 0
        
        length = 2**(N-1)
        
        if k > (length//2):
            rem = k%2
            newk = (k//2) +  rem
            val = self.kthGrammar(N-1,newk)
            if rem == 1:
                return val
            else:
                if val == 1:
                    return 0
                else:
                    return 1
        else:
            return self.kthGrammar(N-1,k)
```

### Approach 2 

```python
class Solution:
    def kthGrammar(self, N: int, K: int) -> int:
        if N == 1:
            return 0
        
        length = 2 ** (N-2)
        
        if K > length:
            if K % length != 0:
                K %= length
            return (self.kthGrammar(N-1, K)) ^ 1
        else:
            return self.kthGrammar(N-1, K)
```
