### Approach 1: DP

```python
class Solution:
    def numTrees(self, n: int) -> int:
        
        G = [None]*(n+1)
        G[0] = 1
        G[1] = 1
        
        for i in range(2,n+1):
            G[i] = 0
            for j in range(1,i+1):
                G[i] = G[i] + (G[j-1] * G[i-j])
                
        return G[n]
```

### Approach 2: Mathematical Deduction

```python
class Solution:
    def numTrees(self, n: int) -> int:
        
        C = 1
        for i in range(0, n):
            C = C * 2*(2*i+1)/(i+2)
        return int(C)
```

```python
class Solution:
    def numTrees(self, n: int) -> int:
        
        G = [None]*(n+1)
        G[0] = 1
        for i in range(1,n+1):
            G[i] = ((2*(2*(i-1) + 1))/((i-1)+2) )*G[i-1]
            
        return int(G[n])
```
