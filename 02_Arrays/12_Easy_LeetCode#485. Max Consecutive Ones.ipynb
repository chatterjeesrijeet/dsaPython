### Approach 1 : 

```python
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:
        
        maxcount = -9999
        tempcount = 0 
        
        for i in range(len(nums)):
            if nums[i] == 1:
                tempcount += 1
            else:
                if tempcount > maxcount:
                    maxcount = tempcount
                tempcount = 0
        
        if tempcount > maxcount:
                    maxcount = tempcount
        
        return maxcount
```
