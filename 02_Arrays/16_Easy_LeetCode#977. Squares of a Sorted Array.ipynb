```python
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        
        for i,ele in enumerate(nums):
            nums[i] = ele*ele
        # print(nums)
        nums.sort()
        # print(nums)
        return nums
```

### Approach 2 : A little More elegant solution 

```python
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        return sorted([x*x for x in nums])
```
