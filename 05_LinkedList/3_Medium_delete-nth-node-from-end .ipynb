```python
class llnode:
    
    def __init__(self,val):
        self.val = val 
        self.next = None
```

```python
def user_input():
    
    input_x  = [int(x) for x in input().strip().split()]
    head = None
    for values in input_x:
        
        if values == -1:
            break
        
        newNode = llnode(values)
        
        if head == None:
            head = newNode
        else:
            temp = head
            while temp.next != None:
                temp = temp.next
            temp.next =  newNode
    return head
```

```python
def printLL(head):
    
    temp = head 
    while temp != None:
        print(temp.val,end = " --> ")
        temp = temp.next
```

```python
head = user_input()
```

```python
printLL(head)
```

```python
def deleNfromEnd(head,n):
    
    if head == None:
        return -1
    
    fut,curr = head, head
    prev = None
    
    while n > 0 :
        fut = fut.next
        n -= 1
    
    while fut != None:
        prev = curr
        curr = curr.next
        fut = fut.next
    
    if prev == None:
        head = head.next
    else:
        prev.next = curr.next
    
    return head
```

```python
head = user_input()
printLL(head)
head = deleNfromEnd(head,3)
print('\n')
printLL(head)
```
