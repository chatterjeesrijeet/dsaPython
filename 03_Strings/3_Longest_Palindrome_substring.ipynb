     Approach 1 :
     
       >> Check for every possible substring and keep the longest one 
       
       Time Complexity :  O(n^3)
       Space Complexity : O(n)

```python
def isPalindrome(st):          #  O(n)
    
    i = 0 
    j = len(st)-1
    
    while i < j :
        if st[i] != st[j]:
            return False
        i += 1
        j -= 1
    return True
```

```python
def longestPalindrome_V1(st):
    lp = ''
    for length in range(1,len(st)+1):
        for i in range(0,len(st)-length+1):          # every possible substring O(n^2)
            tempString  = st[i:i+length]
            if isPalindrome(tempString) and len(tempString) > len(lp):   #isPalindrome : O(n)
                lp = tempString
    return lp
```

```python
print(longestPalindrome_V1('abaxyzzyxf'))
```

     Approach 2 :
     
       >> traverse the string linearly and 
       >> use every index as the center of a palindrom with 
            a) odd length 
            b) even length 
       >> Save the maximum one 
       
       
       Time Complexity :  O(n^2)
       Space Complexity : O(n) --- as we are storing palindromes and at any point in time we might store the biggest palindrome which might be the full string 

              k a t t t a y z z y x f
              
          For i = 3
          
              odd_length_palindrome_check    ===>  k a t <t> t a y z z y x f      o/p : attta
              even_length_palindrome_check   ===>  k a t t <> t a y z z y x f     o/p: tt

```python
def longestPalindromeAtIndexEven(st,idx):
    i = idx
    j = idx + 1
    length = 0
    pst = ''
    while (i >= 0) and (j <= len(st)-1):
        if st[i] == st[j]:
            length += 2
            pst = st[i]+pst+st[j]
            i -= 1 
            j += 1
        else:
            break
    return pst
```

```python
print(longestPalindromeAtIndexEven('abbssbaxyu',3))
print(longestPalindromeAtIndexEven('abbssbaxyu',0))
print(longestPalindromeAtIndexEven('abbssbaxyu',1))
print(longestPalindromeAtIndexEven('abtt',2))
```

```python
def longestPalindromeAtIndexOdd(st,idx):
    i = idx - 1
    j = idx + 1
    length = 0
    pst = st[idx]
    while (i >= 0) and (j <= len(st)-1):
        if st[i] == st[j]:
            length += 2
            pst = st[i]+pst+st[j]
            i -= 1 
            j += 1
        else:
            break
    return pst
```

```python
print(longestPalindromeAtIndexOdd('abbxbssbaxyu',3))
print(longestPalindromeAtIndexOdd('abbssbaxyu',0))
print(longestPalindromeAtIndexOdd('abassbaxyu',1))
print(longestPalindromeAtIndexOdd('abat',3))
```

```python
def longestPalindrome_V2(st):
    lp = ''
    for i in range(len(st)):
        
        lp_odd = longestPalindromeAtIndexOdd(st,i)
        if len(lp_odd) > len(lp):
            lp = lp_odd
        
        lp_even = longestPalindromeAtIndexEven(st,i)
        if len(lp_even)> len(lp):
            lp = lp_even
    return lp
        
```

```python
print(longestPalindrome_V2('abaxyzzyxf'))
```

```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        
        optimallp,optimallen = '',0
        
        for i in range(len(s)):
            
            oddlen , oddlp = 1 , s[i]
            j ,k  = i-1 , i+1
            
            while j>= 0 and k <=len(s)-1:
                if s[j]==s[k]:
                    oddlen += 2
                    oddlp = s[j]+oddlp+s[k]
                    j -= 1
                    k += 1
                else:
                    break
            
            if oddlen > optimallen:
                optimallen , optimallp = oddlen,oddlp
            
            evenlen,evenlp = 0, ''
            j ,k = i, i+1
            
            while j>=0 and k <=len(s)-1:
                if s[j] == s[k]:
                    evenlen += 2
                    evenlp = s[j]+evenlp+s[k]
                    j -= 1
                    k += 1
                else:
                    break
            
            if evenlen > optimallen:
                optimallen, optimallp = evenlen,evenlp
                
        return optimallp
            
            
        
```
