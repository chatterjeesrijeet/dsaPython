```python
import copy
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        
        #base condition 
        if len(nums)==0:
            return [[]]
        
        smallSol = self.subsets(nums[1:])
        
        finalSol = copy.deepcopy(smallSol[:])
        
        
        for ele in smallSol:
            ele.append(nums[0])
            finalSol.append(ele)
                
        return finalSol
```

```python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        n = len(nums)
        output = [[]]
        
        for num in nums:
            output += [curr + [num] for curr in output]
        
        return output
```
