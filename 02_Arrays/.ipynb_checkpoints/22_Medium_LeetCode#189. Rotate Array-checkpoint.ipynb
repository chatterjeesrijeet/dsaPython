```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        
        
        k = k %(len(nums))
        i ,j = 0 , len(nums)-k
        
        while k > 0 :
            temp = nums[len(nums)-1]
            for j in range(len(nums)-1,0,-1):
                nums[j] = nums[j-1]
            nums[0] = temp
            k -= 1
            
        return nums
```

### Approach 2 : 

```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        
        n, a = len(nums) , [0]*n
        
        for i in range(n):
            a[(i+k)%n] = nums[i]
        
        nums[:] = a
        
        return nums
```
