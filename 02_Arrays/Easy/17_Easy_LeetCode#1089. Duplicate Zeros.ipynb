```python
class Solution:
    def duplicateZeros(self, arr: List[int]) -> None:
        """
        Do not return anything, modify arr in-place instead.
        """
        n = len(arr)
        index = 0 
        while index <= n-1:
            if arr[index] == 0:
                for j in range(n-1,index,-1):
                    arr[j] = arr[j-1]
                index += 1
            index += 1
```

### Solution2 : Very Very Tricky 

```python
class Solution:
    def duplicateZeros(self, arr: List[int]) -> None:
        ''' 
        1. need to find the last index from which copying will start 
        2. dupes is basically the count of 0s that doubles and that can give the index 
           from which to start copying 
        '''
        length = len(arr)
        maxval = length - 1
        dupes = 0 
        j = 0
        mydict = {}   #holdes the 0 indexes which will contribute to the doubling 0 
        while j < maxval and j < length:
            if arr[j] == 0:
                maxval -= 1
                dupes += 1
                mydict[j] = 1
            j += 1
        
        #---------- lets now copy ----------------------
        
        if maxval == length-1:
            return
        else:
            
            k = length - 1
            copyfrom = k - dupes
            while k  >= 0:
                if arr[copyfrom] == 0 and mydict.get(copyfrom)==1:
                    arr[k]= 0
                    arr[k-1] = 0
                    k -= 2
                else:
                    arr[k]= arr[copyfrom]
                    k -= 1
                copyfrom -= 1
            
            return
```
