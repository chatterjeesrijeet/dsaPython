### Approach 1 : Using BFS

```python
import queue
class Solution:
    def findCircleNum(self, isConnected: List[List[int]]) -> int:
        
        
        def bfs(sv,visited):
            q = queue.Queue()
            visited[sv] = 1
            q.put(sv)
            
            
            while q.empty() != True:

                poppedNode = q.get()
                
                for i in range(len(visited)):
                    if isConnected[poppedNode][i] == 1 and visited[i] == -1:
                        visited[i] = 1 
                        q.put(i)
            return
                        
        visited = [-1 for i in range(len(isConnected))]
        components = 0 
        for i in range(len(visited)):
            if visited[i] == -1:
                components += 1
                bfs(i,visited)
        
        return components
```

### Approach 2 : Using DFS 

```python
class Solution:
    def findCircleNum(self, isConnected: List[List[int]]) -> int:
        
        '''  --------- Approach 2 --------- '''
        
        def dfs(sv):
            visited[sv] =  1
            for i in range(len(visited)):
                if isConnected[sv][i] == 1 and visited[i] == -1:
                    dfs(i)
            return
                
        visited = [-1 for i in range(len(isConnected))]
        components = 0 
        for i in range(len(visited)):
            if visited[i] == -1:
                components += 1
                dfs(i)
        
        return components
```
