### Approach 1

```python
class Solution:
    def spiralOrder(self, arr: List[List[int]]) -> List[int]:
        
        def findnew(row, col, direction):
            newrow = row
            newcol = col
            if direction == 'right':
                newrow,newcol = row,col+1
            elif direction == 'down':
                newrow,newcol = row+1,col
            elif direction == 'left':
                newrow,newcol = row,col-1
            else:
                newrow,newcol = row-1,col
            return newrow,newcol
        
        directiondict = {'right':'down','down':'left','left':'up', 'up' : 'right'}
        
        visiteddict ={}
        
        nrows , ncols = len(arr), len(arr[0])
        count = 0
        
        
        row, col = 0, 0
        oldrow,oldcol = row,col
        
        direction = 'right'
        ans = []
        
        while count < (nrows*ncols):
            
            if (row,col) not in visiteddict:
                # print(arr[row][col], end = ' ')
                ans.append(arr[row][col])
                visiteddict[(row,col)] = 1
                count += 1
                
            else:
                row,col = oldrow,oldcol
                direction = directiondict[direction]
                
            #compute the new set of indices     
            oldrow,oldcol = row,col
            row, col = findnew(row,col,direction)
            
            #boundry condition 1 :
            if col == ncols:
                row, col = row + 1 , col -1 
                direction = directiondict[direction]
            
            #boundry condition 2 : 
            if row == nrows:
                row, col = row-1, col-1
                direction = directiondict[direction]
            
            # boundry condition 3 :
            if col < 0:
                row, col = row-1, col+1
                direction = directiondict[direction]
                
            
        return ans
```

### Approach 2 :

```python
class Solution:
    def spiralOrder(self, arr: List[List[int]]) -> List[int]:
    
        rows, cols = len(arr) , len(arr[0])
        
        total_ele = rows*cols
        count , itr = 0 , 0 
        i_end , j_end = rows - 1, cols - 1
        
        res = []
        
        while count < total_ele:

            i, j = itr , itr
            
            for j in range(itr,j_end - itr + 1):
                res.append(arr[i][j])
                count += 1
                if count >= total_ele:
                    return res

            for i in range(itr+1,i_end-itr+1):
                res.append(arr[i][j])
                count += 1
                if count >= total_ele:
                    return res

            for j in range(j_end-itr-1,itr - 1,-1):
                res.append(arr[i][j])
                count += 1
                if count >= total_ele:
                    return res

            for i in range(i_end - itr-1,itr,-1):
                res.append(arr[i][j])
                count += 1
                if count >= total_ele:
                    return res

            itr += 1
            
        return res
```
