### Slightly optimal : Although the same concept 

     1. remove unnecessary variables like peakindex we dont need that 
     2. left length and right length compute using the indexes no need so many variables 
        
        Note : Incase of li-ri+1 approach subtract two as li and ri will breakout 
               of while loop with the updated values 

```python
def longestpeak(arr):
    
    mountainLength, i  = 0, 1    

    while i < len(arr) - 1:
        
        # check for peak i.e. index is a peak or not ?
        if arr[i-1] < arr[i] and arr[i+1] < arr[i]:   # peak Check : its a peak if it satisfies the condition  
            
            li = (i-1)
            while li >= 0 and arr[li] < arr[li+1]:
                li -= 1
            
            ri = (i+1)
            while ri < len(arr) and arr[ri] < arr[ri-1]:
                ri += 1
            
            tempMountainLength = ri-li+1-2

            # check and update max length moutain and its peak point
            if tempMountainLength > mountainLength :
                mountainLength = tempMountainLength
                
            
            i = ri
            
        else:
            i +=1

    return mountainLength
```
