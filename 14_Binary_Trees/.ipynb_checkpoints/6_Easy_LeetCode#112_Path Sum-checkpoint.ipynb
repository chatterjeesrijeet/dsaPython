```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        
        if root == None:
            return False
        if root.left == None and root.right == None and root.val == targetSum:
            return True
        
        leftreply = self.hasPathSum(root.left,targetSum-root.val)
        rightreply = self.hasPathSum(root.right,targetSum-root.val)
        
        return leftreply or rightreply
        
```
