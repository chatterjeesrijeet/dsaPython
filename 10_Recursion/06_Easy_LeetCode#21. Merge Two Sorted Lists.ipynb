```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        
        if l1 == None:
            return l2
        
        if l2 == None:
            return l1
        
        H,T = None, None
        
        while l1 != None and l2 != None:
            
            if l1.val < l2.val:
                if H == None:
                    H,T = l1,l1
                else:
                    T.next = l1
                    T = T.next
                l1 = l1.next
            
            elif l2.val <= l1.val:
                if H == None:
                    H,T = l2,l2
                else:
                    T.next = l2
                    T = T.next
                l2 = l2.next
        
        if l1 == None:
            while l2 is not None:
                T.next = l2
                T = T.next
                l2 = l2.next
        
        elif l2 == None:
            while l1 is not None:
                T.next = l1
                T = T.next
                l1 = l1.next
        return H
        
```
