```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
import queue
class Solution:
    
    def levelOrderBottom(self, root: Optional[TreeNode]) -> List[List[int]]:
        
        if root == None:
            return []
        
        q = queue.Queue()
        solution = []
        q.put(root)
        
        while q.empty() != True:
            length = q.qsize()
            level_Sol = []
            while length != 0 :
                node = q.get()
                level_Sol.append(node.val)
                
                if node.left != None:
                    q.put(node.left)
                if node.right != None:
                    q.put(node.right)
                length = length - 1
            solution.append(level_Sol)
            
            
        return solution[::-1]
```
