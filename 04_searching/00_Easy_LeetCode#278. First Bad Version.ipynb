```python
# The isBadVersion API is already defined for you.
# @param version, an integer
# @return an integer
# def isBadVersion(version):

class Solution:
    
    def firstBadVersion(self, n):
        """
        :type n: int
        :rtype: int
        """
        
        return self.helpfirstBadVersion(1,n)
    
    
    def helpfirstBadVersion(self,start,end):
        
        if start > end :
            return -1
        
        mid = (start+end)//2
        
        status = isBadVersion(mid)
        
        if status == False:
            return self.helpfirstBadVersion(mid+1,end)
        else:
            
            leftidx = self.helpfirstBadVersion(start,mid-1)
            if leftidx != -1:
                return leftidx
            return mid
```
