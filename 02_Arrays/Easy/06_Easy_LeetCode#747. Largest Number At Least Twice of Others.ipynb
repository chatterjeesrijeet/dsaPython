```python
class Solution:
    def dominantIndex(self, nums: List[int]) -> int:
        
        if len(nums) == 1:
            return 0
        
        highest_idx = 0
        second_highest_idx = None
        
        for i in range(1,len(nums)):
            if second_highest_idx == None or nums[i] > nums[second_highest_idx]:
                if nums[i] > nums[highest_idx]:
                    second_highest_idx = highest_idx
                    highest_idx = i
                else:
                    second_highest_idx = i
        
        if nums[highest_idx] >= (2*nums[second_highest_idx]):
            return highest_idx 
        else: 
            return -1
```

```python
class Solution:
    def dominantIndex(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return 0       
        n = nums[0:]
        n.sort()
        if n[-1] >= 2 * n[-2]:
            return nums.index(n[-1])    
        return -1
```
