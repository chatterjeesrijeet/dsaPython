```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        
        # numbers.sort()
        i, j = 0 , len(numbers)-1
        
        while i<j :
            tempsum = numbers[i]+numbers[j]
            if tempsum<target:
                i += 1
            elif tempsum > target:
                j -= 1
            else:
                return [i+1,j+1]
        return -1
```
