## Two Cateories  :     
     
     (A) Return the numbers
     
     (B) Return the indices ---> a little tricky as you can not sort the array
     

## Category 1 

### Solution 1 : 

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)-1):
            for j in range(i+1,len(nums)):
                if nums[i]+nums[j] == target:
                    return [i,j]
        return []
```

### Solution 2 : 

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        numdict  = {}
        for i,vals in enumerate(nums):
            remaining = target - vals
            if remaining in numdict:
                return [i,numdict[remaining]]
            else:
                numdict[vals] = i
        return []
```

## Category 2

    Approach 1 : 
        TimeComplexity : O(n^2)
        Space Complexity :    O(1) 

```python
def twoNumSum(arr,x):
    
    for i in range(len(arr)):
        for j in range(len(arr)):
            if i==j:
                continue
            if arr[i]+arr[j] == x:
                return([arr[i],arr[j]])
    return -1    
```

```python
arr = [3,5,-4,8,11,1,-1,6]
x = 10
```

```python
ans = twoNumSum(arr,x)
print(ans)
```

    Sorting & then Double Pointer 
    
    Approach 2 : 
        TimeComplexity : O( nlog(n) + o(n) ) ===> O(nlogn)
        Space Complexity :    O(1) 

```python
def twoNumSum_V2(arr,x):
    
    arr.sort()
    i  = 0
    j = len(arr)-1
    
    while i < j:
        if arr[i] + arr[j] > x:
            j -= 1
        elif arr[i]+arr[j] < x:
            i += 1
        else:
            return [arr[i],arr[j]]
    return -1
```

```python
arr = [3,5,-4,8,11,1,-1,6]
x = 10
```

```python
ans = twoNumSum_V2(arr,x)
print(ans)
```

    Using Hashing: 
        Take a num, look for reminder present in hash table or not !!!
    
    Approach 3 : 
        TimeComplexity : O(n)
        Space Complexity :    O(n) [hastable creation ] 

```python
def twoNumSum_V3(arr,x):
    hashtab = {}
    
    for i in arr:
        reminder = x - i
        if reminder in hashtab:
            return [i,reminder]
        hashtab[i] = True
    return -1
```

```python
arr = [3,5,-4,8,11,1,-1,6]
x = 10
```

```python
ans = twoNumSum_V3(arr,x)
print(ans)
```
