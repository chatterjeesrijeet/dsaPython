#### Approach 1: 

```python
class Solution:
    def reverseWords(self, s: str) -> str:
        
        listofwords = s.strip().split(' ')
        
        newlistofwords= []
        
        for ele in reversed(listofwords):
            if ele != '':
                newlistofwords.append(ele)
        
        
        return ' '.join(newlistofwords)
        
```

### Approach 2 : Simplest

```python
class Solution:
    def reverseWords(self, s: str) -> str:
        return " ".join(reversed(s.split()))
```
