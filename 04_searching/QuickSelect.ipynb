```python
arr = [11,13,1,55,17]
```

```python
def QuickSelect(arr,k):
    k = k - 1
    si,ei = 0 , (len(arr)-1) 
    
    while si <= ei:
        
#         print(f"{si} and {ei}")
        pivotIndex = partition(arr,si,ei)
#         print(arr)
#         print(f"pivot_index = {pivotIndex}")

        if pivotIndex == k :
            return arr[pivotIndex]
        else:
            if pivotIndex < k :
                si = pivotIndex + 1
            else:
                ei = pivotIndex - 1
    
    return -1
```

```python
def partition(arr,si,ei):
    
    pivot_Index = ei
    pivot_Element = arr[ei]
    
    count = 0
    for i in range(si,ei+1):
        if arr[i] < pivot_Element :
            count += 1
    
    correct_position = si + count
    
    arr[pivot_Index],arr[correct_position] = arr[correct_position],arr[pivot_Index]
    
    for i in range(ei,correct_position,-1):
        if arr[i] < pivot_Element:
            for j in range(si,correct_position):
                if arr[j] > pivot_Element:
                    arr[i],arr[j] = arr[j],arr[i]
                    break
    
    return correct_position
```
