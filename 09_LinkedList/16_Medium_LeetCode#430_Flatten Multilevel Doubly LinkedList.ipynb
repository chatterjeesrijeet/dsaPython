```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val, prev, next, child):
        self.val = val
        self.prev = prev
        self.next = next
        self.child = child
"""

class Solution:
    def flatten(self, head: 'Node') -> 'Node':
        
        branchedNode = []
        
        temp = head
        h2 = None
        
        while temp != None or len(branchedNode) != 0:
            
            if temp == None:
                temp = branchedNode.pop()
                temp = temp.next
                
            if temp == None:
                continue
            value = temp.val
            
            newNode = Node(value,None,None,None)
            
            if h2 == None:
                h2 = newNode
                t2 = newNode
            else:
                t2.next = newNode
                newNode.prev = t2 
                t2 = t2.next
            
            if  temp.child != None:
                branchedNode.append(temp)
                temp = temp.child
            else:
                temp = temp.next
        
        return h2
            
        
```
