```python
class linkedlistnode:
    def __init__(self,val):
        self.val = val
        self.next = None

def userinput():
    arr = [int(x) for x in input().strip().split()]
    head,tail = None,None
    for data in arr:
        if data == -1:
            return head
        newNode = linkedlistnode(data)
        if head == None:
            head , tail = newNode , newNode
        else:
            tail.next = newNode
            tail = tail.next
    return head

def printLL(head):
    temp = head 
    while temp != None:
        print(temp.val,end = " --> ")
        temp = temp.next
```

```python
def revLL(head):
    
    while head == None:
        return head
    
    smallhead = revLL(head.next)
    
    if smallhead == None:
        return head
    
    temp,prev = smallhead, None

    while temp != None:
        prev,temp = temp,temp.next
    
    prev.next = head
    head.next = None
    
    return smallhead
    
```

```python
head = userinput()
printLL(head)
print('\n')
head = revLL(head)
printLL(head)
```

```python
def revItr(head):
    
    if head == None or head.next == None:
        return head
    
    past,present,future = None,head,head.next
    
    while present != None:
        present.next = past
        past,present  = present, future
        if future != None:
            future = future.next
    
    head = past
    
    return head
```

```python
head = userinput()
printLL(head)
print('\n')
head = revItr(head)
printLL(head)
```
