```python
class Solution:
    def findDiagonalOrder(self, mat: List[List[int]]) -> List[int]:
        
        def revdir(direction):
            if direction == 1:
                return 0
            else:
                return 1
        
        nrows = len(mat)
        ncols = len(mat[0])
        
        traversedir = 1
        row,col = 0,0
        
        ans = []
        no_ele = 0
        
        while no_ele < (nrows*ncols):
            
            # print(mat[row][col],end = ' ')
            ans.append(mat[row][col])
            
            #--------traverse up and down logic---------------
            if traversedir == 1:
                row , col = row - 1 , col + 1
            else:
                row, col = row + 1 , col - 1
                
            #Boundry condition 1:
            if row < 0 :
                if col == ncols:
                    col , row = col-1, row+2
                else:
                    row += 1
                traversedir = revdir(traversedir)
                    
            #Boundry condition 2:
            if col < 0 :
                if row == nrows:
                    row , col = row-1, col+2 
                else:
                    col += 1
                traversedir = revdir(traversedir)
            
            #Boundry Condition 3 :
            if row == nrows:
                row , col = row-1, col+2
                traversedir = revdir(traversedir)
            
            
            #Boundry Condition 4:
            if col == ncols:
                row , col = row+2 , col-1
                traversedir = revdir(traversedir)
            
            no_ele += 1
            
        return ans
```
