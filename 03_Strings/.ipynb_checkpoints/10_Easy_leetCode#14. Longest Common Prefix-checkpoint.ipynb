### Approach 1 :

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        
        def maxpre(word1,word2):
            k = min(len(word1),len(word2))
            sol = []
            for i in range(k):
                if word1[i]==word2[i]:
                    sol.append(word1[i])
                else:
                    break
            return ''.join(sol)
            
        
        maxprefix = strs[0]
        for i in range(1,len(strs)):
            
            maxprefix = maxpre(maxprefix,strs[i])
            
            
        return maxprefix
```

### Approach 2 : 

```python
class Solution:
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not strs: 
            return ""
        if len(strs) == 1: 
            return strs[0]
        
        strs.sort()
        p = ""
        for x, y in zip(strs[0], strs[-1]):
            if x == y: 
                p+=x
            else: 
                break
        
        return p
```
