```python
import copy
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        
        #base condition 
        if len(nums)==0:
            return [[]]
        
        smallSol = self.subsets(nums[1:])
        
        finalSol = copy.deepcopy(smallSol[:])
        
        
        for ele in smallSol:
            ele.append(nums[0])
            finalSol.append(ele)
                
        return finalSol
```

Algorithm

    Generate all possible binary bitmasks of length n.

    Map a subset to each bitmask: 1 on the ith position in bitmask means the presence of nums[i] in the subset, and 0 means its absence.

    Return output list.

```python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        n = len(nums)
        output = []
        
        for i in range(2**n, 2**(n + 1)):
            # generate bitmask, from 0..00 to 1..11
            bitmask = bin(i)[3:]
            
            # append subset corresponding to that bitmask
            output.append([nums[j] for j in range(n) if bitmask[j] == '1'])
        
        return output
```
