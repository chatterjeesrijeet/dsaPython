```python
class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
        
        solution = []
        
        i = 1
        while i <= numRows:
            # print(solution)
            
            if i == 1:
                solution.append([1])
            
            elif i == 2:
                solution.append([1,1])
            
            else:
                j = i 
                tempsol = []
                
                for j in range(1,i+1):

                    if j == 1 or j == i:
                        tempsol.append(1)
                    else:
                        tempsol.append((solution[i-2][j-2] + solution[i-2][j-1]))
                
                solution.append(tempsol)
                
            i += 1
            
        return solution
```
