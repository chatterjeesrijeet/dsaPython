```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val: int = 0, left: 'Node' = None, right: 'Node' = None, next: 'Node' = None):
        self.val = val
        self.left = left
        self.right = right
        self.next = next
"""
import queue
class Solution:
    def connect(self, root: 'Node') -> 'Node':
        
        
        listofnodes = self.lvl(root)
        print(listofnodes)
        
        for ele in listofnodes:
            j = 0 
            while j < len(ele)-1:
                
                ele[j].next = ele[j+1]
                j += 1
        return root
        
    def lvl(self,root):
        
        if root == None:
            return []
        
        solution = []
        q = queue.Queue()
        q.put(root)
        
        while q.empty()!= True:
            
            length = q.qsize()
            lvlsol = []
            while length != 0:
                node = q.get()
                lvlsol.append(node)
                if node.left != None:
                    q.put(node.left)
                if node.right != None:
                    q.put(node.right)
                    
                length -= 1
            solution.append(lvlsol)
        return solution
```
