```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    
    def countUnivalSubtrees(self, root: Optional[TreeNode]) -> int:
        
        _,count = self.countUnivalSubtreeshelp(root)
        
        return count
        
    def countUnivalSubtreeshelp(self, root):
        
        if root == None:
            return (False,0)
        
        if root.left == None and root.right == None:
            return (True,1)
        
        left_status , leftcount = True , 0
        right_status , rightcount = True , 0
        
        if root.left != None:
            left_status, leftcount  = self.countUnivalSubtreeshelp(root.left)
        
        
        if root.right != None:
            right_status,rightcount = self.countUnivalSubtreeshelp(root.right)
            
        
        if left_status == True and right_status == True:
            rightok,leftok = True,True
            
            if root.right != None:
                rightok = (root.val == root.right.val)
            
            if root.left != None:
                leftok = (root.val == root.left.val)
                
            finalstatus = rightok & leftok 
            
            if finalstatus:
                return ( finalstatus , 1 + leftcount + rightcount )
            else:
                return ( finalstatus , leftcount + rightcount )
                
        
        else:
            
            return (False,leftcount+rightcount)
```
