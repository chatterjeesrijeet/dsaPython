### Approach 1 : 

```python
class Solution:
    def thirdMax(self, nums: List[int]) -> int:
        
        ''' Appraoch 1 '''
        
        maxlist = [None,None,None]
        
        for num in nums:
            if num == maxlist[0] or num == maxlist[1] or num == maxlist[2]:
                continue
            elif maxlist[0] == None or num > maxlist[0]:
                maxlist[1],maxlist[2]= maxlist[0],maxlist[1]
                maxlist[0] = num
            elif maxlist[1]== None or num > maxlist[1]:
                maxlist[2]=maxlist[1]
                maxlist[1]=num
            elif maxlist[2] == None or num > maxlist[2]:
                maxlist[2] = num
#         print(maxlist)
        
        if None in maxlist:
            maxval = None
            for val in maxlist:
                if val == None:
                    pass
                else:
                    if maxval == None:
                        maxval = val
                    else:
                        maxval = max(val,maxval)
            return maxval
        
        else:
            return maxlist[2]
```
