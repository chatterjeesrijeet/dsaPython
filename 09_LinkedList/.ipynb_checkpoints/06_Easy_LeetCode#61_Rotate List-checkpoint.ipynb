```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        
        temp , tail , count  = head , None , 0 
        
        while temp != None:
            count = count + 1
            tail = temp
            temp = temp.next
            
        if count <= 1:
            return head
        
        tail.next = head
        no_of_moves = (count-(k%count))
        prev = None
        while no_of_moves != 0:
            prev = head
            head = head.next
            no_of_moves -= 1
            
        if prev != None:
            prev.next = None
            
        return head
            
        
```
