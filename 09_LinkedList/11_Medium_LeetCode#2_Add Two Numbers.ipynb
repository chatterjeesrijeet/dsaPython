```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        
        i = l1
        j = l2
        
        solutionhead = None
        solutiontail = None
        carry = 0 
        
        while i != None or j != None:
            
            if i == None:
                val1 = 0
            else:
                val1 = i.val
                
            if j == None:
                val2 = 0
            else:
                val2 = j.val
                
            val = (val1 + val2 + carry)%10
            carry =  (val1 + val2+carry)//10
            
            tempNode = ListNode(val)
            if solutionhead == None:
                solutionhead = tempNode
                solutiontail = tempNode
            else:
                solutiontail.next = tempNode
                solutiontail = solutiontail.next
            
            if i != None:
                i = i.next
                
            if j != None:
                j = j.next
        
        if carry != 0 :
            tempNode = ListNode(carry)
            solutiontail.next = tempNode
            solutiontail = solutiontail.next
            
        
        solutiontail.next = None
        
        return solutionhead
```
