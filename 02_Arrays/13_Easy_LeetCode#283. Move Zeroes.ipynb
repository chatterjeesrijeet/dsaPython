### Approach 2:  Two pointer approach 

```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        
        ''' i : index which holds the left most zero
            j : navigates to look for  non-zero ahead of i'''
        
        i = 0  
        # find teh first position with nums[i] == 0
        while i < n:
            if nums[i] ==0 :
                break 
            i = i+1
        
        #initiatlise j as i + 1
        j = i + 1
        # look for first non zero element after i 
        while j < n : 
            if nums[j] != 0 and i < j: 
                nums[i],nums[j] = nums[j],nums[i]
                while nums[i] != 0:
                    i += 1
                    if i == n:
                        return
            j += 1
        return 
```
