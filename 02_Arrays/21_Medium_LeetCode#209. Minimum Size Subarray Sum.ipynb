### Approach 1 : Time Limit Exceeded

```python
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        
        def mysum(i,size):
            
            tsum = 0 
            for j in range(i,i+size):
                tsum = tsum + nums[j]
            # print(i,size,tsum)
            return tsum
        
        size = 1
        
        for size in range(1,len(nums)+1):
            for i in range(len(nums)-size+1):
                tempsum = mysum(i,size)
                if tempsum >= target:
                    return size
        return 0
```

### Approach 2 :  Two pointer approach :

```python
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        
        minterms = 999999999
        left, tempsum = 0 , 0 
        
        for i in range(len(nums)):
            tempsum = tempsum + nums[i]
            while tempsum >= target:
                minterms = min(minterms,i-left+1)
                tempsum = tempsum - nums[left]
                left += 1
        if minterms == 999999999:
            return 0
        else:
            return minterms
```
