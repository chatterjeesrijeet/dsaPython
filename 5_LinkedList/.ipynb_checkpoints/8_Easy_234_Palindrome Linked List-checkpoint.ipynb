```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: ListNode) -> bool:
        
        if head.next == None:
            return True
        
        prev = None
        slow = head
        fast = head
        
        while fast != None and fast.next != None:
            prev = slow
            slow = slow.next
            fast = fast.next.next
        
        if prev != None:
            prev.next = None 
        
        if fast != None:
            slow = slow.next
            
        head2 = slow
        
        # Reverse half of the Linked List 
        i, j = None, head2 
        k = j.next 
        
        while j != None:
            j.next = i 
            i = j 
            j = k 
            if k != None:
                k = k.next
        head2 = i
        
        h1 = head
        h2 = head2 
        
        while h1 != None and h2 != None and h1.val == h2.val:
            h1 , h2 = h1.next, h2.next
        
        if h1 == None and h2 == None:
            return True
        else:
            return False
```
