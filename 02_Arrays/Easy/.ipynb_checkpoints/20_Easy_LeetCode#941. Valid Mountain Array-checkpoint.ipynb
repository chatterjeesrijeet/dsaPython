```python
class Solution:
    def validMountainArray(self, arr: List[int]) -> bool:
        
        length = len(arr)
        if length < 3:
            return False
        
        #---intital up down set up ---
        changeno = 0
        
        for i in range(len(arr)-1):
            
            if (arr[i] < arr[i+1]) and (changeno == 0 or changeno == 1):
                changeno = 1
            
            elif (arr[i] > arr[i+1]) and (changeno == 1 or changeno ==2) :
                changeno = 2 
            
            else:
                return False
                
        if changeno == 2:
            return True
        else:
            return False
```
