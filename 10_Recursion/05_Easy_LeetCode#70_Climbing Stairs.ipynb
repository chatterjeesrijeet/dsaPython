### Approach 1 : DP 

```python
class Solution:
    
    def climbStairs(self, n: int) -> int:
        
        cache = [0]*(n+1)
        
        for i in range(1,n+1):
            
            if i <= 2:
                cache[i] = i
                continue
            
            cache[i] = cache[i-1]+cache[i-2]
        
        return cache[n]
```
