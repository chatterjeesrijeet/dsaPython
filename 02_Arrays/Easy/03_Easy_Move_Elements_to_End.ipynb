```python
def moveToEnd(arr,x):
    
    size = len(arr)
    if size == 0 :
        return 0
    
    #find the position/index of the first replaceable element x
    k = 0 
    while arr[k] != x and k < size:
        k += 1
        if k == size:
            return k
    
    #find the position/index of the first element to be replaced with x from the back 
    j = size - 1
    while arr[j] == x and j > k :
        j -= 1
    
    #now lets keep looping 
    while j > k :
        
        arr[j],arr[k] = arr[k],arr[j]
        j -= 1
        k += 1
        
        while arr[k] != x and k < size:
            k += 1
            
        while arr[j] == x and j > k :
            j -= 1
        
        
    return k
         
```
