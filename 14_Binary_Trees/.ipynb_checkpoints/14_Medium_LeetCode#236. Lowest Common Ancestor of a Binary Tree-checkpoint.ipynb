### Approach 1 : Recursive Approach

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        
        if root == None:
            return False 
        
        myStatus , leftStatus , rightStatus = False, False, False
        
        if root.val == p.val or root.val == q.val :
            myStatus = True
        
        leftStatus = self.lowestCommonAncestor(root.left,p,q)
        rightStatus = self.lowestCommonAncestor(root.right,p,q)
        
        if (rightStatus == True and myStatus == True) or  (leftStatus == True and myStatus == True) or (rightStatus == True and leftStatus == True):
            return root
        else:
            return (myStatus or leftStatus or rightStatus)
```
