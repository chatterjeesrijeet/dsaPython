### Approach 1 

```python
class Solution:
    def validPalindrome(self, s: str) -> bool:
        
        if s == s[::-1]:
            return True 
        
        for i in range(len(s)):
            newstr = s[:i] + s[i+1:]
            if newstr == newstr[::-1]:
                return True
        return False
```

```python
class Solution:
    def validPalindrome(self, s: str) -> bool:
        
        def ispalindrome(l):
            strlist = ''.join(l)
            return strlist == strlist[::-1]
        
        pallist = list(s)
        
        i , j = 0 , len(pallist)-1
        
        while i < j :
            if pallist[i] != pallist[j]:
                return ( ispalindrome(pallist[:i]+pallist[i+1:]) or ispalindrome(pallist[:j]+pallist[j+1:]) )
            i , j = i+1, j-1
        
        
        return True
```

```python
class Solution:
    
    def validPalindrome(self, s: str) -> bool:
        
        pallist = list(s)
        
        i , j = 0 , len(pallist)-1
        
        while i < j :
            if pallist[i] != pallist[j]:
                return ( self.checkpalindrome(s[:i]+s[i+1:]) or self.checkpalindrome(s[:j]+s[j+1:]) )
            i , j = i+1, j-1
        
        
        return True
    
    def checkpalindrome(self,s):
        if s[::-1] == s:
            return True

        return False
```
