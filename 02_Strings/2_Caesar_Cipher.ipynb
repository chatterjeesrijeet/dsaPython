            Time Complexity : O(n)
            Space Complexity : O(n)    #as you are creating a string of same length n

```python
def CaesarCipher_V1(myString,X):
    shiftedStr = []
    
    for char in myString:
        base_number = (ord(char) - 97  + X)%26
        ascii_val = base_number + 97
        shiftedStr.append(chr(ascii_val))
    
    return ''.join(shiftedStr)
```

```python
print(CaesarCipher_V1('a',26))
print(CaesarCipher_V1('aed',2))
print(CaesarCipher_V1('ayz',2))
```
