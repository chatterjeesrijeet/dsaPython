### Approach 1 :  Naive Approach

```python
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:
        
        ''' Approach 1 : Time Limit Exceeded'''
    
        if not(0 in nums):
            return len(nums)
    
        def maxones(arr):
            count = 0
            i = 0 
            while i < len(arr):
                
                if arr[i]==1:
                    j = i 
                    while j < len(arr) and arr[j]==1:
                        j+=1
                    tempcount = (j-i)
                    if tempcount > count:
                        count = tempcount
                    i = j
                else:
                    i += 1
            return count
        
        val = 0
        for idx in range(len(nums)):
            
            if nums[idx] == 0:
                temparr = nums[:]
                temparr[idx] = 1
                tempval = maxones(temparr)
                if tempval>val:
                    val = tempval
                    
        return val
```
