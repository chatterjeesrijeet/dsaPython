```python
class Solution:
    def reverseWords(self, s: str) -> str:
        
        listofwords = s.strip().split(' ')
        
        newlistofwords= []
        
        for ele in reversed(listofwords):
            if ele != '':
                newlistofwords.append(ele)
        
        
        return ' '.join(newlistofwords)
        
```
