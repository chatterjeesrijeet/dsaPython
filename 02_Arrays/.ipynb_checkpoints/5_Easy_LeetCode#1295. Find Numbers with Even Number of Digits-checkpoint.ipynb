### Approach 1 :

```python
class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        count = 0
        for ele in nums:
            if len(str(ele))%2 == 0:count += 1
        return count
```

### Approach 2 : 

```python
class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        
        count = 0 
        
        for number in nums:
            
            digits = 0 
            
            while number > 0:
                digits += 1
                number //= 10
            
            if digits %2 == 0:
                count += 1
        
        return count 
            
```
