```python
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        i,j = 0, 0
        n1,n2 = m,n
        
        k = 0 
        arr1 = nums1[:]
        
        while i < n1 and j < n2:
            if arr1[i] < nums2[j]:
                nums1[k] = arr1[i]
                i += 1
            else:
                nums1[k] = nums2[j]
                j += 1
            k += 1
        
        if i == n1:
            while j != n2:
                nums1[k] = nums2[j]
                j += 1
                k += 1
        
        if j == n2:
            while i != n1:
                nums1[k] = arr1[i]
                i += 1
                k += 1
        
        return
```
