```python
class Solution:
    
    def countPrimes(self, n: int) -> int:
        
        if n <= 2:
            return 0

        primecount = [0,0] + [1] * (n-2)
        i = 2
        while i * i < n :
            if primecount[i]:
                primecount[i*i : n : i] = [0] * ( (n - i*i -1 ) // i + 1 )
            i += 1 if i == 2 else 2
        return sum(primecount)
```
