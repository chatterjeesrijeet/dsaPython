```python
class Solution:
    def removeElement(self, arr: List[int], x: int) -> int:
        
        if len(arr) == 0:
            return 0
        
        i , j = 0 , len(arr)-1
        
        while i < j :
            
            while arr[i] != x and i < j :
                i += 1
            
            while arr[j] == x and j > i :
                j -= 1
            
            arr[i], arr[j]  = arr[j], arr[i]
            
        if arr[j] == x:
            return i
        else:
            return i+1
```
