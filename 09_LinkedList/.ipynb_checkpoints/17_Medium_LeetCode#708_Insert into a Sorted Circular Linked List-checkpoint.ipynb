```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val=None, next=None):
        self.val = val
        self.next = next
"""

class Solution:
    def insert(self, head: 'Node', insertVal: int) -> 'Node':
        
        '''scenario 1 : empty list''' 
        if not head :
            newNode = Node(insertVal)
            head = newNode
            newNode.next = newNode
            return head
        
        ''' scenario 2 : only one node in the list''' 
        
        p1 = head 
        p2 = head.next
        
        newNode = Node(insertVal)
        
        if p1==p2:
            p1.next = newNode
            newNode.next = p1
            return head
        
        '''scenario 3 : more than max value or less than min value'''
        
        #find the max val and min value and their pointer 
        
        maxVal = head.val
        maxValP = head
        
        minVal = head.val
        minValP = head
        
        temp = head.next 
        
        while temp != head:
            if temp.val >= maxVal:
                maxVal = temp.val
                maxValP = temp
            elif temp.val < minVal:
                minVal = temp.val
                minValP = temp
            temp = temp.next
        
        if insertVal >= maxVal:
            newNode.next = maxValP.next
            maxValP.next = newNode
            return head
        
        elif insertVal <= minVal:
            temp = head 
            while temp.next  != minValP:
                temp = temp.next
            temp.next = newNode
            newNode.next = minValP
            return head
            
        ''' scenario 4 : inbetweek two numbers as per number line present in the list '''    
        while not(insertVal >= p1.val and insertVal <= p2.val):
            p1 = p2
            p2 = p2.next
        
        p1.next = newNode
        newNode.next = p2
        
        return head
```
