    (a) Duplicacy Check:If immediate pop hits an open parenthesis ‘(‘, then we have found a duplicate parenthesis. 
    
    For example, (((a+b))+c) has duplicate brackets around a+b. When we reach second “)” after a+b, we have “((” in the stack. Since the top of stack is an opening bracket, we conclude that there are duplicate brackets. 
 
 
 
    (b) Not Duplicate but Unnecessary :  If immediate pop doesn’t hit any operand(‘*’, ‘+’, ‘/’, ‘-‘) then it indicates the presence of unwanted brackets surrounded by expression. For instance, (a)+b contain unwanted () around a thus it is redundant. 

```python
def checkredundancy(s):
    
    stack = []
    
    for char in s:
        
        if char == ')':
            topChar = stack[len(stack)-1]
            
            if topChar == '(':
                return True
            
            countOp = 0
            while topChar != '(':
                if topChar == '+' or topChar=='-' or topChar=='*' or topChar=='/':
                    countOp += 1
                stack.pop()
                topChar = stack[len(stack)-1]
            
            if countOp == 0:
                return True
            else:
                stack.pop()
                
        else:
            stack.append(char)
    return False
```

```python
checkredundancy('((a+b))')
```

```python
checkredundancy('(a+(b)/c)')               #this case is not duplicate but redundant 
```

```python
checkredundancy('(a+b*(c-d))')
```
