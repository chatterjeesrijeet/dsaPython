```python
def findgroupanagrams(l):
    myHash ={}
    solution = []
    for words in l:
        
        charlist = [s for s in words]
        charlist.sort()
        sortedWord = ''.join(charlist)
        
        if sortedWord in myHash:
            myHash[sortedWord].append(words)
        else:
            myHash[sortedWord] = [words]
    
    for values in myHash.values():
        solution.append(values)
    
    return solution
```

```python
l = ['yo','act','flop','tac','cat','oy','olfp']
solution = findgroupanagrams(l)
print(solution)
```
