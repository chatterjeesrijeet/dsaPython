```python
class linkedlistnode:
    def __init__(self,val):
        self.val = val
        self.next=None
        
def userinput():
    
    arr = [int(val) for val in input().strip().split()]
    
    head , tail = None, None
    
    for val in arr:
        
        if val == -1:
            break
        
        newNode = linkedlistnode(val)
        
        if head == None:
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = tail.next
    
    return head

def printLL(head):
    temp = head 
    while temp != None:
        print(temp.val,end = ' --> ')
        temp = temp.next
```
