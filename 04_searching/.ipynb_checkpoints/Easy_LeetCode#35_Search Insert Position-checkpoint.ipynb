```python
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        
        
        si = 0 
        ei = len(nums)-1
        
        while si<= ei:
            mid = (si+ei)//2
            print(mid)
            
            if nums[mid] == target:
                return mid
            
            elif target > nums[mid]:
                si = mid+1
            else:
                ei = mid -1 
        return si
        
```
