### Approach 1 : DP 

```python
class Solution:
    
    def climbStairs(self, n: int) -> int:
        
        cache = [0]*(n+1)
        
        for i in range(1,n+1):
            
            if i <= 2:
                cache[i] = i
                continue
            
            cache[i] = cache[i-1]+cache[i-2]
        
        return cache[n]
```

### Approach 2 : Fibbonachi Number

```python
class Solution:
    
    def climbStairs(self, n: int) -> int:
        # Appraoch 2
        if n <= 2:
            return n
        
        first , second = 1 , 2
        for i in range(3,n+1):
            third = first + second
            first = second
            second = third
        
        return second
```

### Appraoch 3 : (N+1)th Fbbonacchi Number 

```python
class Solution:
    
    def climbStairs(self, n: int) -> int:
        # Approach 3 :
        
        golden_ratio = (1+(5**0.5))/2
        
        return int(round((golden_ratio**(n+1))/(5**0.5)))
```
