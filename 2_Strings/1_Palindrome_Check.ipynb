    Appraoch 1: 
    
            Time Complexity : O(n^2)
            Space Complexity : O(n)
        

```python
def isPalindrome_V1(st):
    
    reverserdSt = ''
    
    for i in range(len(st)-1,-1,-1):       # O(n)   
        reverserdSt += st[i]     #this creates a new string everytime i.e. copies old chars then adds st[i] to it 
                                 # this makes it O(n)
    '''
    for i in reversed(range(len(st))):
        reverserdSt += st[i] 
    '''
    return st==reverserdSt
```

```python
print(isPalindrome_V1('a'))
print(isPalindrome_V1('ab'))
print(isPalindrome_V1('aba'))
```

    **Appraoch 2:** 
            >> No need to create a new string all the time instead save it in list 
            
            Time Complexity : O(n)
            Space Complexity : O(n)

```python
def isPalindrome_V2(st):
    
    reverserdSt = []
    
    for i in range(len(st)-1,-1,-1):       # O(n)   
        reverserdSt.append(st[i])
    
    return st==''.join(reverserdSt)
```

```python
print(isPalindrome_V2('a'))
print(isPalindrome_V2('ab'))
print(isPalindrome_V2('aba'))
```

    Approach 3: 
    
            Time Complexity : O(n^2)
            Space Complexity : O(n)
        

```python
def isPalindrome_V3(st):
    return st == st[::-1]
```

```python
print(isPalindrome_V3('a'))
print(isPalindrome_V3('ab'))
print(isPalindrome_V3('aba'))
```

    Appraoch 4:  Best : Two Pointer Approach 
    
            Time Complexity : O(n)
            Space Complexity : O(1)
        

```python
def isPalindrome_V4(st):
    i = 0
    j = len(st)-1
    
    while i < j :
        if st[i]!=st[j]:
            return False
        i += 1
        j -= 1
    return True
            
```

```python
myString = 'abcdcba'
ans = isPalindrome_V4(myString)
print(ans)
```

```python
print(isPalindrome_V4('a'))
print(isPalindrome_V4('ab'))
print(isPalindrome_V4('aba'))
```
