                                           ...... Approach 1 ......

```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        l1 = list(word1)
        l2 = list(word2)
        
        i = 0
        finallist = []
        while i < len(l1) and i < len(l2):
            finallist.append(l1[i])
            finallist.append(l2[i])
            i += 1
        
        if len(l1)==len(l2):
            return ''.join(finallist)
        elif i == len(l1):
            finallist.extend(l2[i:])
        else:
            finallist.extend(l1[i:])
            
        return ''.join(finallist)
```

                                           ...... Approach 2 ......


    1. rather than appending into a list ...just create a list of fixed size and insert at index 

    2. appending will end up doubling the list and copy pasting

```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        l1 = list(word1)
        l2 = list(word2)
        
        len1 = len(l1)
        len2 = len(l2)
        
        i,index  = 0, 0
        finallist = [None] * (len(l1)+len(l2))
        
        while i < len(l1) and i < len(l2):
            finallist[index] = l1[i]
            finallist[index+1] = l2[i]
            i += 1
            index += 2
        
        
        if len(l1)==len(l2):
            ''.join(finallist)
        
        elif i == len(l1):
            while i < len(l2):
                # print(index)
                finallist[index] = l2[i]
                i += 1
                index += 1
                
        else:
            while i < len(l1):
                finallist[index] = l1[i]
                i += 1
                index += 1
            
        return ''.join(finallist)
```

                                        ..... Approach 3 .....

     no need to convert the input strings into lists as you can directly fine length and subscript by index 

```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
    
        len1 = len(word1)
        len2 = len(word2)
        
        i,index  = 0, 0
        finallist = [None] * (len1+len2)
        
        while i < len1 and i < len2:
            finallist[index] = word1[i]
            finallist[index+1] = word2[i]
            i += 1
            index += 2
        
        
        if len1==len2:
            ''.join(finallist)
        
        elif i == len1:
            while i < len2:
                # print(index)
                finallist[index] = word2[i]
                i += 1
                index += 1
                
        else:
            while i < len1:
                finallist[index] = word1[i]
                i += 1
                index += 1
            
        return ''.join(finallist)
```
