```python
class linkedlistnode:
    def __init__(self,val):
        self.val = val
        self.next=None
        
def userinput():
    
    arr = [int(val) for val in input().strip().split()]
    
    head , tail = None, None
    
    for val in arr:
        
        if val == -1:
            break
        
        newNode = linkedlistnode(val)
        
        if head == None:
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = tail.next
    return head

def printLL(head):
    temp = head 
    while temp != None:
        print(temp.val,end = ' --> ')
        temp = temp.next
```

```python
head = userinput()
printLL(head)
```

```python
def partitionlist(head,x):
    
    if head == None or head.next == None:
            return head
        
    #find teh position where to switch 
    posprev, pos = None, head
    while pos != None and pos.val < x:
        posprev = pos
        pos = pos.next

    if pos == None:
        return head

    #now traverse the list to find elements greater than x
    tempprev, temp = posprev, pos

    while temp != None:

        if temp.val < x :

            tempprev.next = temp.next

            if posprev != None:
                posprev.next = temp
                temp.next = pos
                posprev = posprev.next
            else:
                temp.next = pos
                head = temp
                posprev = temp
            temp = tempprev.next

        else:
            tempprev = temp
            temp = temp.next
    return head
```

```python
head = userinput()
printLL(head)
print('\n')
head = partitionlist(head,3)
printLL(head)
```
