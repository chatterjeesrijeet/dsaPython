```python
class Solution:
    def fib(self, n: int) -> int:
        def fibhelp(n,fibnumbers):
            if n == 0 :
                return 0
            if n == 1:
                return 1
            
            if fibnumbers[n-1] == None:
                fibnumbers[n-1] = fibhelp(n-1,fibnumbers)
            if fibnumbers[n-2] == None:
                fibnumbers[n-2] = fibhelp(n-2,fibnumbers)
            
            return fibnumbers[n-1]+fibnumbers[n-2]
        
        fibnumbers = [None]*(n+1)
        
        return fibhelp(n,fibnumbers)
```
