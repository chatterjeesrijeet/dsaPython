
        >> Approach 1 : you will have the minimum and maximum only once unless minimum or maximum element has appeared twice.
        
              1. While pushing in self.stack, you push to the self.minMaxStack, iff the new element is either  
                 
                  (a) less than equal to curr minimum or 
                  (b) greater than or equal to current maximum
              
              2. This way your space will be optimized in minMaxStack but the processing time will be a little higher side during pop()
              
              3. pop() time you must check the deleted element is either current minimum or maximum then pop() from self.minMaxStack else no need
              
        >> Appraoch 2 : 
              
               1. you keep pushing the minimum and maximum as many elements you push in self.stack ....
               2. In that case you have duplicate min and max pushed on top of each other 
               3. but pop() fucntion become svery easy pop means pop() from both minMaxStack and from stack.
    
    

### Approach 1

```python
class MinMaxStack:
    
    def __init__(self):
        self.minMaxStack = [] # insert in the form of tuples ,list or dict ..I choose dict
        self.stack = []
        
    def peek(self):
        size = len(self.stack)
        if size==0:
            return -1
        return self.stack[size-1]
    
    def push(self,val):
        size = len(self.stack)
        
        if len(self.stack) == 0:
            self.minMaxStack.append({'minimum' : val , 'maximum' : val})
        
        else:
            idx = len(self.minMaxStack)
            currMin = self.minMaxStack[idx-1]['minimum']
            currMax = self.minMaxStack[idx-1]['maximum']
            
            if val <= currMin:
                self.minMaxStack.append({'minimum' : val , 'maximum' : currMax})
            elif val >= currMax:
                self.minMaxStack.append({'minimum' : currMin , 'maximum' : val})
        
        self.stack.append(val)
        
    def pop(self):
        
        size = len(self.stack)
        
        if size == 0:
            return -1
        else:
            
            poppedVal = self.stack[size-1]
            
            idx = len(self.minMaxStack)
            currMin = self.minMaxStack[idx-1]['minimum']
            currMax = self.minMaxStack[idx-1]['maximum']
            
            if poppedVal == currMin or poppedVal == currMax:
                self.minMaxStack.pop()
            
            self.stack.pop()
            
    def getMin(self):
        
        size = len(self.stack)
        
        if len(self.stack) == 0:
            return -1
        
        else:
            idx = len(self.minMaxStack)
            currMin = self.minMaxStack[idx-1]['minimum']
            return currMin
    
    def getMax(self):
        
        size = len(self.stack)
        
        if len(self.stack) == 0:
            return -1
        
        else:
            idx = len(self.minMaxStack)
            currMax = self.minMaxStack[idx-1]['maximum']
            return currMax
        
    def __str__(self):
        return 'Stack : ' + str(self.stack) + '\n\n' +'minMax Stack : ' + str(self.minMaxStack)
```
