```python

import queue

class Graph:
    
    def __init__(self,n):
        self.n = n 
        self.mat = [[-1 for i in range(n)] for j in range(n)]
        for i in range(n):
            self.mat[i][i] = 1
    
    def addEdge(self,v1,v2):
        self.mat[v1][v2] = 1
        self.mat[v2][v1] = 1
    
    def removeEdge(self,v1,v2):
        self.mat[v1][v2]= -1
        self.mat[v2][v1] = -1
        
    def noOfProvinces(self):
        
        visited = [-1 for i in range(self.n)]
        noOfProvinces = 0
        
        for i in range(len(visited)):
            if visited[i] != 1:
                noOfProvinces += 1
                self.bfsTraversal(visited,sv = i)
                
        return noOfProvinces
                
    def bfsTraversal(self,visited,sv):
        
        q = queue.Queue()
        q.put(sv)
        visited[sv] =1 
        
        while q.empty() != 1:
            
            popped = q.get()
            
            for i in range(len(self.mat[popped])):
                
                if self.mat[popped][i] == 1 and visited[i] != 1:
                    q.put(i)
                    visited[i] = 1
        
        
        
```
