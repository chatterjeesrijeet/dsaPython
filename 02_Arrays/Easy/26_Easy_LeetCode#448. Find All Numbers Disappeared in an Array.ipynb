```python
class Solution:
    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:
        
        mydict = {}
        
        for val in nums:
            if mydict.get(val) == 1:
                pass
            else:
                mydict[val] = 1
            
        solution = []
        
        for i in range(1,len(nums)+1):
            if mydict.get(i) == 1:
                pass
            else:
                solution.append(i)
                
                
        return solution
```
