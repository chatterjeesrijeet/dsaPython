```python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        
        if n < 0 :
            return 1/self.myPowhelp(x,abs(n))
        return self.myPowhelp(x,n)
    
    def myPowhelp(self,x,n):
        
        if n == 1:
            return (x)
        if n == 0:
            return 1
        
        if n % 2 == 0 :
            mid = self.myPowhelp(x,n//2)
            return mid*mid
        
        else:
            mid = self.myPowhelp(x,n//2)
            return (mid*mid)*x
```
