```python
class StockSpanner:

    def __init__(self):
        self.st = []
        self.prices = []

    def next(self, price: int) -> int:
        
        if len(self.prices) == 0:
            self.st.append(0)
            self.prices.append(price)
            return 1
        
        else:
            self.prices.append(price)
            
            i = len(self.prices) - 1
            top_index = self.st[-1]
            
            if self.prices[top_index] > self.prices[i]:
                self.st.append(i)
                return (i-top_index)
            else:
                while len(self.st) > 0 and self.prices[top_index] <= self.prices[i]:
                    self.st.pop()
                    if len(self.st) > 0:
                        top_index = self.st[-1]
                    
                if len(self.st)==0:
                    self.st.append(i)
                    return i + 1
                else:
                    span = i - self.st[-1]
                    self.st.append(i)
                    return span
            

# Your StockSpanner object will be instantiated and called as such:
# obj = StockSpanner()
# param_1 = obj.next(price)
```
