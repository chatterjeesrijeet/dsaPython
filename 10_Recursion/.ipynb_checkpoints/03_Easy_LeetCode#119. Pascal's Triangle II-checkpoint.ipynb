```python
class Solution:
    def getRow(self, rowIndex: int) -> List[int]:
        sol = []
        
        mat = [[-99 for col in range(rowIndex)]for row in range(rowIndex)]
        
        for j in range(rowIndex+1):
            sol.append(self.getRowHelp(rowIndex,j,mat))
        
        return sol
        
    def getRowHelp(self,row,col,mat):
        
        if col==row or col == 0:
            return 1
        
        if mat[row-1][col-1] == -99:
            mat[row-1][col-1] = self.getRowHelp(row-1,col-1,mat)
            
        if mat[row-1][col] == -99:
            mat[row-1][col] = self.getRowHelp(row-1,col,mat)
        
        return  mat[row-1][col-1] + mat[row-1][col]
```
