```python
class StackNode:
     
    def __init__(self, data):
        self.data = data
        self.next = None
 
class Stack:
     
    def __init__(self):
        self.top = None
      
    def push(self, data):
        if (self.top == None):
            self.top = StackNode(data)
            return
        s = StackNode(data)
        s.next = self.top
        self.top = s
      
    def pop(self):
        s = self.top
        self.top = self.top.next
        return s
  
    def display(self):
        s = self.top
        while (s != None):
            print(s.data, end = ' ')
            s = s.next
         
    # Reverses the stack using simple
    # linked list reversal logic.
    
    def reverse(self):
        
        prev,curr = None,self.top
        fut = curr.next
        
        curr.next = prev
        
        while fut != None:
            prev = curr
            curr = fut
            fut = fut.next
            curr.next = prev
        
        self.top = curr
```

```python
s = Stack()
s.push(1)
s.push(2)
s.push(3)
s.push(4)

print("Original Stack")
s.display()
print()

# Reverse
s.reverse()

print("Reversed Stack")
s.display()
```
